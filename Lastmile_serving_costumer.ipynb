```python
#Lectura de archivo tipo .csv
Lastmile_serving_costumer = pd.read_csv('lastmile_delivery_operations_mit_lift_lab_serving_customer.csv')
Lastmile_serving_costumer
```

```python
#Filtro de columnas para cualitativas
Cualitativa = Lastmile_serving_costumer.iloc[:,[0,1,2,3,5,6,7,8,13,17,21,22,23,25]]
```

```python
#Filtro de columnas para cuantitativas
Cuantitativa = Lastmile_serving_costumer.iloc[:,[4,9,11,12,14,15,16,18,19,20,24,26]]
```

```python
#Corroboramos valores nulos
valores_nulos = Cualitativa.isnull().sum()
valores_nulos 
```

```python
#Limpiamos Valores nulos de cualitativos
Cualitativa = Cualitativa.fillna(method='ffill')
```

```python
#Corroboramos valores nulos de Cualitativos para limpiarlos
valores_nulos = Cualitativa.isnull().sum()
valores_nulos
```

```python
#Limpiamos los valores nulos de Cuantitativas
Cuantitativa = Cuantitativa.fillna(round(Cuantitativa.mean(),1))
```

```python
#Corroboramos valores nulos de Cualitativos para limpiarlos
valores_nulos_cuantitativa = Cuantitativa.isnull().sum()
valores_nulos_cuantitativa
```

```python
#Método de IQR
y = Cuantitativa
percentile25 = y.quantile(0.25) #Q1
percentile75 = y.quantile(0.75) #Q3
iqr = percentile75 - percentile25

Limite_Superior_iqr = percentile75 + 1.5*iqr
Limite_Inferior_iqr = percentile25 - 1.5*iqr
print("Limite superior permitido", Limite_Superior_iqr)
print("Limite inferior permitido", Limite_Inferior_iqr)
```

```python
#Obtenemos datos limpios
data_clean_iqr = Cuantitativa[(y <= Limite_Superior_iqr)&(y >= Limite_Inferior_iqr)]
data_clean_iqr
```

```python
#Unimos ambos dataframes ya que estan limpios
datos_limpios_Lastmile_serving_costumer = pd.concat([Cuantitativa,Cualitativa],axis=1)
datos_limpios_Lastmile_serving_costumer
```

```python
#Convertir dataframe a CSV
datos_limpios_Lastmile_serving_costumer.to_csv("datos_limpios_Lastmile_serving_costumer.csv")
#Descargar archivo filtrado en csv
from google.colab import files


files.download("datos_limpios_Lastmile_serving_costumer.csv")
```
